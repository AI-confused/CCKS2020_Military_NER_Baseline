```python
import json
import pandas as pd
import os
import numpy as np
from numpy import  *
import torch
import tqdm as tqdm
```

```python
path_pre=['../output/roberta_large_baseline']
K_FOLD_FROM=0
K_FOLD_TO=5
Total_fold=(K_FOLD_TO-K_FOLD_FROM)*len(path_pre)
not_used=[]*len(path_pre)
file_name=['test_logit_res.json']
#添加信息
Message='roberta_softmax_5fold'
test_sample_dir='./data/ccks_8_data_v2'
saved_model_path=path_pre[0]+str(0)
assert len(Message)>0,'add info!'

```

```python
all_json_dict=[]
for ipathpre,ifilename in zip(path_pre,file_name):
    for ifold in range(K_FOLD_FROM,K_FOLD_TO):
        filename=os.path.join(ipathpre+str(ifold),ifilename)
        with open(filename,'r') as f:
            json_dict=json.load(f)
        print(json_dict)
        all_json_dict.append(json_dict['res'])
```

```python
#获取每个样本的testid
test_id=[]
for  i_index in range(len(all_json_dict[0])):
    for i_test_id in all_json_dict[0][i_index].keys():
        test_id.append(i_test_id)
test_id_sub_key=[]
for  i_test_id_sub_key in (all_json_dict[0][0][test_id[0]]):
    test_id_sub_key.append(i_test_id_sub_key)
```

```python
#ensemble
final_dict_json={}
#融合key对应的logit
ensemble_key=['ipre_seq_logit']
for i_samp_index,asample_testid in enumerate(test_id):
    final_dict_json[asample_testid]={}
    for i_test_id_sub_key in test_id_sub_key:
        if i_test_id_sub_key not in ensemble_key and i_test_id_sub_key not in  final_dict_json[asample_testid].keys():
            final_dict_json[asample_testid][i_test_id_sub_key]=\
            all_json_dict[0][i_samp_index][asample_testid][i_test_id_sub_key]
            continue
        if i_test_id_sub_key  in ensemble_key:
            for ifold in range(0,Total_fold):
                if ifold in not_used:
                    continue
                print('ifold:',ifold)

                ijson_dict=all_json_dict[ifold]

                #print(list(final_dict_json[iexample].keys()))
                if i_test_id_sub_key not in list(final_dict_json[asample_testid].keys()):
                    final_dict_json[asample_testid][i_test_id_sub_key]=np.array(ijson_dict[i_samp_index][asample_testid][i_test_id_sub_key])/\
                    (1.0*(Total_fold-len(not_used)))
                    
                else:

                    def ensemble_pred(final_dict_json,ijson_dict):            
                        return final_dict_json[asample_testid][i_test_id_sub_key]+\
                    np.array(ijson_dict[i_samp_index][asample_testid][i_test_id_sub_key])/(1.0*(Total_fold-len(not_used)))

                    final_dict_json[asample_testid][i_test_id_sub_key]\
                    =ensemble_pred(final_dict_json,ijson_dict)




```

```python
def change_ilabel(label_list):
    for i_index,ilabel in enumerate(label_list):
        if ilabel>4:
            label_list[i_index]-=4
    return label_list

def get_a_sentence_label(test_res,save_list,ipre_seq_label,true_len,org_text,token2orgindex,id2label):
    # 抽离出有效位置对应的文本

    #保存字典（测试集需要）
    a_senten_res=[]

    ipre_seq_label = ipre_seq_label[:true_len][1:-1]
    previous_label_id = 0
    point_start = -1
    #print(ipre_seq_label)
    #处理label的值，便于split
    ipre_seq_label=change_ilabel(ipre_seq_label)
    for i_pos, ipos_pre_label in enumerate(ipre_seq_label):

        if ipos_pre_label != previous_label_id:
            # 起始点
            if previous_label_id == 0:
                point_start = i_pos
            else:
                tmp_dict={}
                tmp_dict['label_type']=id2label[previous_label_id]
                tmp_dict['overlap']=0
                tmp_dict['start_pos']=token2orgindex[point_start]+1
                #包括end
                tmp_dict['end_pos']=token2orgindex[i_pos]
                a_senten_res.append(tmp_dict)
                save_list.append(
                id2label[previous_label_id] + '_' + org_text[token2orgindex[point_start]:token2orgindex[i_pos]])
                point_start = i_pos

        previous_label_id = ipos_pre_label

        if i_pos == len(ipre_seq_label) - 1 and ipos_pre_label != 0:
            tmp_dict = {}
            tmp_dict['label_type'] = id2label[previous_label_id]
            tmp_dict['overlap'] = 0
            tmp_dict['start_pos'] = token2orgindex[point_start] + 1
            # 包括end
            tmp_dict['end_pos'] = token2orgindex[i_pos]
            a_senten_res.append(tmp_dict)
            save_list.append(id2label[ipos_pre_label] + '_' + org_text[token2orgindex[point_start]:])
    test_res.append(a_senten_res)
    
```

```python
#使用融合后的logit计算结果
final_test_res=[]
save_list=[]
for itestid,ires_dict in final_dict_json.items():
      #单个样本信息
    token2orgindex = ires_dict['token2orgindex']
    org_text = ires_dict['org_text']
    true_len = ires_dict['true_len']
    id2label={int(id_):label for id_,label in (ires_dict['id2label']).items()}
    
    #print((id2label))
    ipre_seq_logit=ires_dict['ipre_seq_logit']
    ipre_seq_label = np.argmax(ipre_seq_logit, axis=1).tolist()
    get_a_sentence_label(final_test_res,save_list,ipre_seq_label,true_len,org_text,token2orgindex,id2label)

    
```

```python
#打开提交样例
with open(os.path.join(test_sample_dir,'submit_sample.json'), 'r', encoding='utf-8') as f:
    json_dict = json.load(f)
    json_keys=list(json_dict.keys())
    for ijson_key,isen_res in zip(json_keys,final_test_res):
        json_dict[ijson_key]=isen_res
#保存结果
with open(os.path.join(saved_model_path,'test_res'+Message+'.json'),'w',encoding='utf-8') as f:
    json.dump(json_dict,f,ensure_ascii=False)

```
